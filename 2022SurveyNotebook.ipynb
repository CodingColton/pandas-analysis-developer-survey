```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('data/stack-overflow-developer-survey-2022/survey_results_public.csv')
```

```python
schema_df = pd.read_csv('data/stack-overflow-developer-survey-2022/survey_results_schema.csv')
```

## Clean Data:

#### Drop rows with NAN in EdLevel

```python
df = df[df['EdLevel'].notna()]
```

## Charts

### Question: What education level do most developers have?

```python
education = [education for education, df in df.groupby('EdLevel')]
results = df.groupby('EdLevel').count()

plt.bar(education, results['ResponseId'])
plt.xticks(education, rotation='vertical', size=8)
plt.ylabel('Total Count')
plt.xlabel('Education Type')
    
plt.show()
```
